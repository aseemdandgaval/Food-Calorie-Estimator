## **Making the dataset for Regression**

### **Make datasets for the following foods**

* Banana
* Fried dough twist
* Bun
* Grape

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### **Banana**

```python
# For Bananas

label_path = 'ECUSTFD-resized-/Regression/banana/'
banana_labels = []
x_banana = []
y_banana = []

files = [f for f in os.listdir(label_path) if os.path.isfile(os.path.join(label_path, f))]

for file in files:
    # construct the full label and image path
    file_path = os.path.join(label_path, file)
    
    # open the file and read its contents
    with open(file_path, 'r') as f:
        contents = f.read()
        data = contents.split('\n')
        line_data = []
        
        for i in range(len(data)-1):
            line_data.append(data[i].split(' '))
        
        if line_data[0][0] != '0':
            x_scale_factor = 25/(float(line_data[1][3]))
            y_scale_factor = 25/(float(line_data[1][4]))
            denorm_x = x_scale_factor * float(line_data[0][3])
            denorm_y = y_scale_factor * float(line_data[0][4])
            x_banana.append(round(denorm_x, 3))
            y_banana.append(round(denorm_y, 3))
            banana_labels.append(file.rsplit('.', 1)[0])
        else:
            x_scale_factor = 25/(float(line_data[0][3]))
            y_scale_factor = 25/(float(line_data[0][4]))
            denorm_x = x_scale_factor * float(line_data[1][3])
            denorm_y = y_scale_factor * float(line_data[1][4])
            x_banana.append(round(denorm_x, 3))
            y_banana.append(round(denorm_y, 3))
            banana_labels.append(file.rsplit('.', 1)[0])


#print("")
#print(banana_labels)
#print("")
#print(x_banana)
#print("")
#print(y_banana)
#print(len(y_banana))

df = pd.DataFrame({'labels': banana_labels, 'x': x_banana, 'y': y_banana})
df.to_csv('regression_banana_data.csv_raw', index=False)
df.head(10)
```
